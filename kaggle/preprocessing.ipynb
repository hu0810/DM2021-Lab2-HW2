```python
import pandas as pd
import numpy as np
```

```python
data_path = "/NTHU/DATA MINING/lab2_kaggle/data"
```

```python
#load data 
tweets_df = pd.read_json(data_path + '/tweets_DM.json',lines=True)

data_identification_df = pd.read_csv(data_path + "/data_identification.csv")

emotions_df = pd.read_csv(data_path + "/emotion.csv")



tweets_df['text'] = tweets_df['_source'].apply(lambda x:x['tweet']['text'])
tweets_df['id'] = tweets_df['_source'].apply(lambda x:x['tweet']['tweet_id'])
tweets_df['hashtags'] = tweets_df['_source'].apply(lambda x:x['tweet']['hashtags'])
```

```python
tweets_df
```

```python
emotions_df = emotions_df.rename(index=str,columns={"tweet_id":"id"})
emotions_df
```

```python
data_identification_df = data_identification_df.rename(index=str,columns={"tweet_id":"id"})
data_identification_df
```

```python
tweets_df = pd.merge(tweets_df,data_identification_df, on="id")
train_df = tweets_df[tweets_df["identification"] == "train"]
test_df = tweets_df[tweets_df["identification"] == "test"]
train_df
```

```python
test_df
```

```python
train_df = pd.merge(train_df,emotions_df, on="id")
```

```python
train_df
```

```python
train_df = train_df.drop(columns = ['identification','_score','_index','_source','_crawldate','_type'])
test_df = test_df.drop(columns = ['identification','_score','_index','_source','_crawldate','_type'])
```

```python
train_df.set_index("id",inplace=True)
```

```python
test_df.set_index("id",inplace=True)
```

```python
test_df = test_df.drop(columns = ['hashtags'])
```

```python
test_df
```

```python
train_df
```

```python
possible_labels = train_df.emotion.unique()

label_dict = {}
for index, possible_label in enumerate(possible_labels):
    label_dict[possible_label] = index
label_dict
```

```python
train_df['label'] = train_df.emotion.replace(label_dict)
```

```python
train_df = train_df.drop(columns = ['hashtags','label','data_type'])
```

```python
train_df.to_pickle("/NTHU/DATA MINING/lab2_kaggle/training_pkl" )
```

```python
df = pd.read_csv("/NTHU/DATA MINING/lab2_kaggle/training_csv")

s = pd.Series(list(df['emotion']))
one_hot_emotion = pd.get_dummies(s)
```

```python
text_bert = df[['id', 'text']]
```

```python
train_df = pd.concat([text_bert, one_hot_emotion.reindex(text_bert.index)], axis=1)
```
